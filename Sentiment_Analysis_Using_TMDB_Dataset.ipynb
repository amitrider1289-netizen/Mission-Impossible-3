Before a single model is trained, we perform a deep-tissue scan of the IMDB corpus to identify the linguistic patterns that define human sentiment. Our EDA is governed by three pillars: *Distribution, **Composition, and **Correlation*.

### 🔍 *EDA Objectives & Methodology*
* ⚖️ *Statistical Distribution:* Verifying the 1:1 class balance and analyzing *Review Length Variance* to see if length correlates with sentiment.
* 🔤 *Textual Composition:* Performing *N-Gram Analysis* (Unigrams/Bigrams) and assessing *Stop-Word Impact* to reveal the core vocabulary of cinema critique.
* ☁️ *Sentiment Correlation:* Utilizing *Word Clouds* and measuring *Lexical Diversity* to identify if "Elite" reviewers use different language patterns than casual viewers.

```python
import numpy as np

# ROI (Return on Investment) calculation
# We use .replace and .fillna to handle movies with 0 budget
df['roi'] = ((df['revenue'] - df['budget']) / df['budget']).replace([np.inf, -np.inf], np.nan).fillna(0)

# Vote ratio to understand popularity relative to engagement
df['vote_ratio'] = df['vote_average'] / (df['vote_count'] + 1)

# Extraction of decade for historical trends
df['release_decade'] = (df['release_year'] // 10) * 10

# Create a binary target for classification if needed (Hit or Flop)
df['hit_flop'] = (df['revenue'] > df['budget']).astype(int)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Numeric Pipeline: Fill missing values with median and scale
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Categorical Pipeline: Fill missing with 'most_frequent' and encode
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

# Combine into a single preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd

# 1. Define models with "Fast" settings
models = {
    "Linear Regression": LinearRegression(),
    # Limit depth and trees to speed up baseline
    "Random Forest": RandomForestRegressor(n_estimators=50, max_depth=10, n_jobs=-1, random_state=42),
    # Limit boosting rounds for speed
    "XGBoost": XGBRegressor(n_estimators=50, max_depth=5, n_jobs=-1, random_state=42)
}

results = []

print("Starting Baseline Evaluation (Optimized)...")

for name, model_obj in models.items():
    try:
        print(f"Training {name}...")

        pipeline = Pipeline(steps=[
            ('preprocessor', preprocessor),
            ('regressor', model_obj)
        ])

        pipeline.fit(X_train, y_train)
        y_pred = pipeline.predict(X_test)

        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        r2 = r2_score(y_test, y_pred)

        results.append({"Model": name, "RMSE": round(rmse, 4), "R2": round(r2, 4)})
        print(f"Finished {name}.")

    except Exception as e:
        print(f"!!! Error with {name}: {e}")

if results:
    print("\n--- Baseline Results ---")
    print(pd.DataFrame(results))
```
